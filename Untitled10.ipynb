```python
import os, shutil

SRC = "/content/Dataset Of Animal Images/Dataset Of animal Images"  # go one level deeper
DST = "/content/Animals_Clean"

print("SRC exists?", os.path.exists(SRC))

# Reset clean dir
if os.path.exists(DST):
    shutil.rmtree(DST)
os.makedirs(DST, exist_ok=True)

def is_image(f):
    return f.lower().endswith((".jpg", ".jpeg", ".png"))

report = {}
for cls in sorted(os.listdir(SRC)):
    cls_path = os.path.join(SRC, cls)
    if not os.path.isdir(cls_path):
        continue

    candidates = [
        os.path.join(cls_path, "train", "images"),
        os.path.join(cls_path, "images"),
        os.path.join(cls_path, "train"),
        cls_path
    ]
    found = []
    for p in candidates:
        if os.path.isdir(p):
            for f in os.listdir(p):
                if is_image(f):
                    found.append(os.path.join(p, f))

    report[cls] = found

# Print summary
print("\n=== Scan Summary ===")
for cls, imgs in report.items():
    print(f"{cls}: {len(imgs)} images")
print("====================\n")

# Copy images
for cls, imgs in report.items():
    if len(imgs) == 0:
        continue
    dst_cls = os.path.join(DST, cls)
    os.makedirs(dst_cls, exist_ok=True)
    for src_img in imgs[:500]:  # copy only first 500 per class to avoid memory issues
        shutil.copy(src_img, dst_cls)

print("✅ Dataset cleaned successfully!")

# Show results
for cls in os.listdir(DST):
    files = os.listdir(os.path.join(DST, cls))[:3]
    print(f"{cls}: sample files {files}")
```

```python
# Colab cell: TFLite model inspector + simple image inference with labels
# Run after you've uploaded your model via files.upload()

# Install optional support lib for reading TFLite metadata
!pip install -q tflite-support Pillow

from google.colab import files
import os, io, sys, glob
from PIL import Image
import numpy as np
import tensorflow as tf

# --- 1) Upload files: model.tflite (and optionally labels.txt and test images) ---
print("Upload your .tflite model file (and optionally labels.txt and test images).")
uploaded = files.upload()   # use the picker to upload model.tflite and labels/test images

# Find the tflite filename
tflite_fname = None
label_fname = None
image_files = []

for name in uploaded.keys():
    if name.lower().endswith(".tflite"):
        tflite_fname = name
    if name.lower().endswith(".txt") and 'label' in name.lower():
        label_fname = name
    if name.lower().endswith((".jpg", ".jpeg", ".png")):
        image_files.append(name)

print("Found tflite:", tflite_fname)
print("Found label file:", label_fname)
print("Found image files:", image_files)

if tflite_fname is None:
    raise SystemExit("Please upload a .tflite file and run this cell again.")


# --- 2) Print interpreter input/output details (small 'summary') ---
interpreter = tf.lite.Interpreter(model_path=tflite_fname)
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("\n=== MODEL SUMMARY ===")
print("Inputs:")
for i, inp in enumerate(input_details):
    print(f"  [{i}] name: {inp.get('name')}, shape: {inp.get('shape')}, dtype: {inp.get('dtype')}")
print("Outputs:")
for i, out in enumerate(output_details):
    print(f"  [{i}] name: {out.get('name')}, shape: {out.get('shape')}, dtype: {out.get('dtype')}")


# --- 3) Try to read labels ---
labels = None

# Option A: if user uploaded a labels file like labels.txt
if label_fname:
    with open(label_fname, "r", encoding="utf-8") as f:
        labels = [line.strip() for line in f.readlines() if line.strip()]
    print(f"\nLoaded {len(labels)} labels from {label_fname}")

# Option B: try to read metadata labels (if labels not uploaded)
if labels is None:
    try:
        from tflite_support import metadata
        displayer = metadata.MetadataDisplayer.with_model_file(tflite_fname)
        # get associated files (e.g., label file)
        assoc_files = displayer.get_packed_associated_file_list()
        if assoc_files:
            # read first associated file that looks like labels
            for fn in assoc_files:
                if "label" in fn.lower() or fn.lower().endswith(".txt"):
                    raw = displayer.get_associated_file_buffer(fn)
                    labels = raw.decode("utf-8").splitlines()
                    print(f"\nLoaded {len(labels)} labels from model metadata file: {fn}")
                    break
    except Exception as e:
        # metadata lib may fail for some models; ignore and continue
        #print("Metadata read failed:", e)
        pass

if labels is None:
    print("\nNo labels found automatically. If you have a labels.txt, upload it or create a list of class names.")
    labels = None
else:
    # show label examples
    for i,lab in enumerate(labels[:10]):
        print(f"  {i}: {lab}")


# --- 4) Helper: preprocess image to match model input ---
def preprocess_image(img_path, input_shape, input_dtype):
    """
    Load an image, resize to input_shape (H,W), and convert dtype.
    Adjust normalization depending on model (try float 0-1 if dtype=float32).
    """
    img = Image.open(img_path).convert("RGB")
    h, w = input_shape[1], input_shape[2]  # expecting shape [1,h,w,3]
    img = img.resize((w, h))
    arr = np.asarray(img)

    if np.issubdtype(input_dtype, np.floating):
        # common: scale to [0,1]
        arr = arr.astype(np.float32) / 255.0
    else:
        # quantized models often expect uint8 0-255
        arr = arr.astype(input_dtype)

    # expand dims to batch
    arr = np.expand_dims(arr, axis=0)
    return arr

# Determine primary input detail (assumes single input)
in_det = input_details[0]
in_shape = in_det['shape']  # e.g. [1,224,224,3]
in_dtype = in_det['dtype']

# --- 5) If no images uploaded, ask to upload test images now ---
if len(image_files) == 0:
    print("\nNo test images uploaded yet. Use the file picker to upload images you want to test now.")
    more = files.upload()
    for name in more.keys():
        if name.lower().endswith((".jpg", ".jpeg", ".png")):
            image_files.append(name)
    print("Now found images:", image_files)

if len(image_files) == 0:
    print("No images provided. You can still inspect model input/output above.")
else:
    # --- 6) Run inference on each test image and print top predictions ---
    print("\n=== RUNNING INFERENCE ON TEST IMAGES ===")
    for imgf in image_files:
        try:
            input_data = preprocess_image(imgf, in_shape, in_dtype)
            # If input dtype is uint8 but we provided float, convert:
            if in_dtype == np.uint8 and input_data.dtype != np.uint8:
                input_data = (input_data * 255).astype(np.uint8)

            interpreter.set_tensor(in_det['index'], input_data)
            interpreter.invoke()
            output_data = interpreter.get_tensor(output_details[0]['index'])

            # handle various output shapes (probabilities or logits)
            # if output is [1, N] take first row
            if output_data.ndim == 2 and output_data.shape[0] == 1:
                scores = output_data[0]
            elif output_data.ndim == 1:
                scores = output_data
            else:
                # try to flatten
                scores = output_data.flatten()

            # softmax if not probabilities and many classes
            if np.max(scores) > 1.0 or np.min(scores) < 0.0:
                # try to softmax
                exp = np.exp(scores - np.max(scores))
                probs = exp / exp.sum()
            else:
                probs = scores

            top_k = probs.argsort()[-5:][::-1]  # top 5 indices
            print(f"\nImage: {imgf}")
            for idx in top_k:
                prob = probs[idx]
                name = labels[idx] if (labels is not None and idx < len(labels)) else f"Class_{idx}"
                print(f"  {name}: {prob:.4f}")
        except Exception as e:
            print(f"Error processing {imgf}: {e}")

# --- 7) Optional: print short summary output (classes found among test images) ---
if len(image_files) > 0 and labels is not None:
    detected = {}
    for imgf in image_files:
        try:
            input_data = preprocess_image(imgf, in_shape, in_dtype)
            if in_dtype == np.uint8 and input_data.dtype != np.uint8:
                input_data = (input_data * 255).astype(np.uint8)
            interpreter.set_tensor(in_det['index'], input_data)
            interpreter.invoke()
            output_data = interpreter.get_tensor(output_details[0]['index'])
            scores = output_data[0] if output_data.ndim==2 and output_data.shape[0]==1 else output_data.flatten()
            probs = (np.exp(scores - np.max(scores)) / np.sum(np.exp(scores - np.max(scores)))) if (np.max(scores) > 1.0 or np.min(scores) < 0.0) else scores
            idx = int(np.argmax(probs))
            name = labels[idx] if idx < len(labels) else f"Class_{idx}"
            detected[name] = detected.get(name, 0) + 1
        except:
            pass

    if detected:
        print("\n=== SUMMARY OF PREDICTIONS ON PROVIDED IMAGES ===")
        for k,v in detected.items():
            print(f"  {k}: {v} image(s)")
```

```python
# Colab cell: Bulk inference & summary for a folder-structured dataset using a .tflite model
# 1) Install pillow (usually present), tensorflow (Colab already has tf)
!pip install -q Pillow

from google.colab import files
import tensorflow as tf
import numpy as np
from PIL import Image
import os, io, zipfile, csv, sys, glob

# -------------- 1) Upload model.tflite and dataset zip (or upload dataset folders manually) --------------
print("Upload your model (.tflite).")
uploaded = files.upload()   # upload model.tflite first
tflite_fname = None
for name in uploaded.keys():
    if name.lower().endswith(".tflite"):
        tflite_fname = name

if tflite_fname is None:
    raise SystemExit("No .tflite found. Upload the tflite model and rerun.")

print("Now upload your dataset zip (optional) or upload dataset folders directly.")
more = files.upload()  # upload dataset zip or images/folders
dataset_root = "dataset_from_upload"
# create dataset folder
if not os.path.exists(dataset_root):
    os.makedirs(dataset_root)

# If user uploaded a zip file, extract it. Otherwise, uploaded file structure may already contain folders.
for name in more.keys():
    if name.lower().endswith(".zip"):
        print("Extracting", name)
        with zipfile.ZipFile(name, 'r') as z:
            z.extractall(dataset_root)
    else:
        # if images or folders uploaded directly, save them under dataset_root
        # uploaded dict contains bytes; write files preserving names
        with open(os.path.join(dataset_root, name), "wb") as f:
            f.write(more[name])
        # Note: direct folder uploads via Colab UI may not preserve folder structure. Prefer zip.

# If extraction created nested folder like 'Dataset Of animal Images', try to find the top-level folder
# Find first directory inside dataset_root; if it's a single directory, use it
entries = [os.path.join(dataset_root, e) for e in os.listdir(dataset_root)]
dirs = [p for p in entries if os.path.isdir(p)]
if len(dirs) == 1:
    dataset_root = dirs[0]
print("Dataset root used:", dataset_root)

# -------------- 2) Walk dataset folders to collect image file paths and true labels --------------
# We expect structure: dataset_root/ClassName/*.jpg
image_paths = []
true_labels = []
class_names = sorted([d for d in os.listdir(dataset_root) if os.path.isdir(os.path.join(dataset_root, d))])

if len(class_names) == 0:
    raise SystemExit("No class folders detected inside dataset root. Check the zip structure.")

print("Detected classes:", class_names)

for cls in class_names:
    cls_dir = os.path.join(dataset_root, cls)
    for ext in ("*.jpg","*.jpeg","*.png","*.bmp"):
        for img in glob.glob(os.path.join(cls_dir, ext)):
            image_paths.append(img)
            true_labels.append(cls)

print(f"Found {len(image_paths)} images across {len(class_names)} classes.")

# -------------- 3) Load TFLite model and get input/output details --------------
interpreter = tf.lite.Interpreter(model_path=tflite_fname)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# assume single input
in_det = input_details[0]
in_shape = in_det['shape']  # example [1,h,w,3]
in_dtype = in_det['dtype']
# normalize decision: if float32 -> scale to [0,1], else keep uint8
is_float_model = np.issubdtype(in_dtype, np.floating)

print("\nModel input shape:", in_shape, "dtype:", in_dtype)
print("Model output shape:", output_details[0]['shape'], "dtype:", output_details[0]['dtype'])
height, width = int(in_shape[1]), int(in_shape[2])

# -------------- 4) Helper: preprocess image --------------
def preprocess_image(path):
    img = Image.open(path).convert("RGB")
    img = img.resize((width, height))
    arr = np.array(img)
    if is_float_model:
        arr = arr.astype(np.float32) / 255.0
    else:
        arr = arr.astype(np.uint8)
    arr = np.expand_dims(arr, axis=0)  # add batch dim
    return arr

# -------------- 5) Run inference over all images, collect results --------------
preds = []
probs_list = []
pred_labels = []
correct = 0

# If you don't have a labels.txt, map indices to class_names in sorted order.
# Warning: The model output ordering must match this mapping. If you saved labels during training, use that labels file.
label_map = class_names  # index 0->class_names[0], etc.

for img_path, true in zip(image_paths, true_labels):
    try:
        input_data = preprocess_image(img_path)
        # adjust dtype if needed
        if not is_float_model and input_data.dtype != np.uint8:
            input_data = (input_data * 255).astype(np.uint8)
        interpreter.set_tensor(in_det['index'], input_data)
        interpreter.invoke()
        out = interpreter.get_tensor(output_details[0]['index'])
        # flatten to 1D scores
        if out.ndim == 2 and out.shape[0] == 1:
            scores = out[0]
        else:
            scores = out.flatten()
        # convert to probabilities if needed
        if scores.max() > 1.0 or scores.min() < 0.0:
            exp = np.exp(scores - scores.max())
            probs = exp / exp.sum()
        else:
            probs = scores
        top_idx = int(np.argmax(probs))
        pred = label_map[top_idx] if top_idx < len(label_map) else f"Class_{top_idx}"
        prob = float(probs[top_idx])
        preds.append(pred)
        probs_list.append(prob)
        pred_labels.append(top_idx)
        if pred == true:
            correct += 1
    except Exception as e:
        print("Error on", img_path, ":", e)
        preds.append("ERROR")
        probs_list.append(0.0)
        pred_labels.append(-1)

# -------------- 6) Print summary and per-class counts --------------
total = len(image_paths)
acc = correct / total if total>0 else 0.0
print(f"\nTotal images: {total}")
print(f"Correct predictions: {correct}")
print(f"Overall accuracy: {acc*100:.2f}%\n")

# per-class breakdown
from collections import Counter, defaultdict
per_true = defaultdict(int)
per_correct = defaultdict(int)
per_predcount = defaultdict(int)
for t,p,pred_label in zip(true_labels, preds, pred_labels):
    per_true[t] += 1
    per_predcount[p] += 1
    if p == t:
        per_correct[t] += 1

print("Per-class results:")
for cls in class_names:
    total_cls = per_true.get(cls,0)
    corr = per_correct.get(cls,0)
    predc = per_predcount.get(cls,0)
    acc_cls = (corr/total_cls*100) if total_cls>0 else 0
    print(f"  {cls}: {total_cls} images, correct={corr}, accuracy={acc_cls:.2f}% , predicted_total={predc}")

# -------------- 7) Save CSV of predictions --------------
csv_fname = "predictions.csv"
with open(csv_fname, "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["image_path","true_label","pred_label","pred_prob"])
    for img_path, true, pred, prob in zip(image_paths, true_labels, preds, probs_list):
        writer.writerow([img_path, true, pred, f"{prob:.6f}"])
print("\nSaved predictions to", csv_fname)
files.download(csv_fname)
```
