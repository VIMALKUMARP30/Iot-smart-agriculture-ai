```python

!git clone https://github.com/ultralytics/yolov5
%cd yolov5
!pip install -r requirements.txt gdown

file_id = "14EsvQiRy7z1ZR7g8ldExIXCTi-JaW-3x"
!gdown --id {file_id} -O dataset.zip


!rm -rf dataset
!unzip -q dataset.zip -d dataset
!ls dataset

import os, shutil, random, yaml

root = "dataset"
subs = os.listdir(root)
if len(subs) == 1 and os.path.isdir(os.path.join(root, subs[0])):
    os.rename(os.path.join(root, subs[0]), os.path.join(root, "animals"))
    dataset_folder = "dataset/animals"
else:
    dataset_folder = "dataset"

print("Using dataset folder:", dataset_folder)

for split in ["train/images", "train/labels", "val/images", "val/labels"]:
    os.makedirs(os.path.join(dataset_folder, split), exist_ok=True)

all_images = []
for root_dir, _, files in os.walk(dataset_folder):
    for f in files:
        if f.lower().endswith((".jpg", ".png", ".jpeg")):
            all_images.append(os.path.join(root_dir, f))

print(f"Found {len(all_images)} images")

random.shuffle(all_images)
split_idx = int(0.8 * len(all_images))
train_imgs, val_imgs = all_images[:split_idx], all_images[split_idx:]

def move_files(img_list, split):
    for img in img_list:
        label = os.path.splitext(img)[0] + ".txt"
        # Move image
        shutil.copy(img, os.path.join(dataset_folder, f"{split}/images"))
        # Move label if exists
        if os.path.exists(label):
            shutil.copy(label, os.path.join(dataset_folder, f"{split}/labels"))

move_files(train_imgs, "train")
move_files(val_imgs, "val")

print("✅ Train/Val split done")

yaml_path = os.path.join(dataset_folder, "data.yaml")
classes = set()
for f in os.listdir(os.path.join(dataset_folder, "train/labels")):
    if f.endswith(".txt"):
        for line in open(os.path.join(dataset_folder, "train/labels", f)):
            if line.strip():
                classes.add(int(line.split()[0]))
classes = sorted(list(classes))
names = [f"class_{i}" for i in classes]
data_yaml = {
    "train": os.path.join(dataset_folder, "train/images"),
    "val": os.path.join(dataset_folder, "val/images"),
    "nc": len(names),
    "names": names
}
with open(yaml_path, "w") as f:
    yaml.dump(data_yaml, f)

print("✅ data.yaml created:")
print(open(yaml_path).read())


import os
os.environ["WANDB_MODE"] = "disabled"

!python train.py --img 320 --batch 16 --epochs 50 --data {yaml_path} --weights yolov5n.pt --cache


!python export.py --weights runs/train/exp/weights/best.pt --include tflite --img 320


from google.colab import files
files.download("runs/train/exp/weights/best.tflite")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import files
import matplotlib.pyplot as plt

TRAIN_DIR = "/content/Animals_Clean"
IMG_SIZE = 128
BATCH = 32
EPOCHS = 20


datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

train_gen = datagen.flow_from_directory(
    TRAIN_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH, subset="training", shuffle=True, seed=42
)
val_gen = datagen.flow_from_directory(
    TRAIN_DIR, target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH, subset="validation", shuffle=False, seed=42
)

print("✅ Classes:", train_gen.class_indices)

base = tf.keras.applications.MobileNetV2(
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    weights="imagenet", include_top=False, pooling="avg"
)


for layer in base.layers:
    layer.trainable = False

x = tf.keras.layers.Dense(128, activation="relu")(base.output)
out = tf.keras.layers.Dense(train_gen.num_classes, activation="softmax")(x)

model = tf.keras.Model(inputs=base.input, outputs=out)

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])


model.fit(train_gen, validation_data=val_gen, epochs=5)

for layer in base.layers[-40:]:
    layer.trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(1e-5),
              loss='categorical_crossentropy', metrics=['accuracy'])


history = model.fit(train_gen, validation_data=val_gen, epochs=EPOCHS)


def rep_dataset():
    for imgs, _ in train_gen:
        yield [imgs.astype("float32")]
        break

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = rep_dataset

try:
    tflite_model = converter.convert()
except Exception:
    converter.representative_dataset = None
    tflite_model = converter.convert()

tflite_path = "animal_classifier.tflite"
with open(tflite_path, "wb") as f:
    f.write(tflite_model)

print("✅ Saved:", tflite_path)
files.download(tflite_path)

=
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.legend(); plt.title("Accuracy"); plt.show()

plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(); plt.title("Loss"); plt.show()
```

```python

import os
import numpy as np
import tensorflow as tf
from PIL import Image
import sys

TFLITE_PATH = "animal_classifier.tflite"
REPRESENTATIVE_DATA_DIR = "dataset_clean_for_quant"
QUANT_OUTPUT = "animal_classifier_int8.tflite"

def human_size(bytesize):
    for unit in ['B','KB','MB','GB']:
        if bytesize < 1024.0:
            return f"{bytesize:.1f}{unit}"
        bytesize /= 1024.0
    return f"{bytesize:.1f}TB"

def inspect_tflite(path):
    print(f"\nInspecting: {path}")
    if not os.path.exists(path):
        print("ERROR: file not found:", path)
        return None
    try:
        interpreter = tf.lite.Interpreter(model_path=path)
        interpreter.allocate_tensors()
    except Exception as e:
        print("ERROR loading TFLite model:", e)
        return None

    input_details = interpreter.get_input_details()
    output_details = interpreter.get_output_details()

    print("Model file size:", human_size(os.path.getsize(path)))
    print("\n-- Inputs:")
    for i,inp in enumerate(input_details):
        print(f" Input #{i}: name={inp.get('name')}, shape={list(inp['shape'])}, dtype={inp['dtype']}")
        q = inp.get('quantization', ())
        print("   quantization:", q)
    print("\n-- Outputs:")
    for i,out in enumerate(output_details):
        print(f" Output #{i}: name={out.get('name')}, shape={list(out['shape'])}, dtype={out['dtype']}")
        q = out.get('quantization', ())
        print("   quantization:", q)

    info = {}
    info['input_shape'] = list(input_details[0]['shape'])
    info['input_dtype'] = str(input_details[0]['dtype'])
    info['input_quant'] = input_details[0].get('quantization', (0.0,0))
    info['output_shape'] = list(output_details[0]['shape'])
    info['output_dtype'] = str(output_details[0]['dtype'])
    info['size_bytes'] = os.path.getsize(path)
    return info

def recommend_for_esp32(info):
    print("\n=== ESP32-CAM compatibility check (simple heuristic) ===")
    ok = True
    reasons = []

    ishape = info['input_shape']
    if len(ishape) != 4:
        ok = False
        reasons.append("Input tensor not 4D (expected [1,H,W,3])")
    else:
        _,h,w,c = ishape
        if c != 3:
            reasons.append(f"Warning: input channels = {c} (expected 3 RGB)")
        if max(h,w) > 160:
            ok = False
            reasons.append(f"Input resolution {h}x{w} too large for ESP32-CAM; prefer 96 or 128.")
        if h != w:
            reasons.append("Prefer square input (e.g., 128x128) for performance and simple resizing.")

    dtype = info['input_dtype']
    if 'uint8' in dtype or 'int8' in dtype:
        reasons.append(f"Quantized input dtype = {dtype} (good for ESP32).")
    else:
        ok = False
        reasons.append(f"Input dtype = {dtype} (float model). Recommend INT8/uint8 quantization for ESP32.")

    size_mb = info['size_bytes'] / (1024*1024)
    if size_mb > 3.0:
        ok = False
        reasons.append(f"Model size {size_mb:.2f} MB — large for ESP32. Try smaller model or aggressive quantization.")
    else:
        reasons.append(f"Model size {size_mb:.2f} MB")

    if len(info['output_shape']) != 2 and not (len(info['output_shape'])==1 or len(info['output_shape'])==2):
        reasons.append("Output shape is unusual; expected [1,num_classes] or [num_classes].")

    if ok:
        print("-> Looks OK for ESP32-CAM (assuming PSRAM or small model).")
    else:
        print("-> NOT READY for ESP32-CAM. Need changes before deployment.")
    for r in reasons:
        print(" -", r)
    return ok

def convert_float_to_int8(orig_tflite, image_folder, out_path, input_size=(128,128)):
    print("\nConverting to INT8 (post-training quantization) ...")
    print("NOTE: Best quantization requires original SavedModel / Keras model. If you only have float .tflite, please re-export from training (SavedModel) then run this cell.")
    return False

if _name_ == "_main_":
    info = inspect_tflite(TFLITE_PATH)
    if info is None:
        print("\nCannot inspect model. Make sure TFLITE_PATH is correct.")
        sys.exit(1)

    ok = recommend_for_esp32(info)

    print("\nNext steps recommendations:")
    if ok:
        print(" - You can convert .tflite -> C header and deploy to ESP32-CAM. (Use xxd or Python script to create model_data.h).")
    else:
        # give advice
        if 'float' in info['input_dtype']:
            print(" - Model is float. Re-export/convert to INT8/uint8 with representative dataset (post-training quantization).")
            print("   (I can give Colab code to quantize if you have original SavedModel or Keras model.)")
        if info['size_bytes']/(1024*1024) > 3.0:
            print(" - Model is large. Try smaller backbone or prune/quantize.")
        if max(info['input_shape'][1], info['input_shape'][2]) > 160:
            print(" - Resize model input (e.g., retrain with 128x128 or 96x96).")
    print("\nIf you want: I can also provide a Colab cell to CONVERT a SavedModel/Keras model to uint8 TFLite (with representative dataset).")
    print("Or, I can provide the Python script to create model_data.h from your .tflite for Arduino upload.")
```
